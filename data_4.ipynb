```python
dataset =np.random.randn(25)
sns.rugplot(dataset)
```

```python
plt.hist(dataset,alpha=0.5)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min=dataset.min()-2
x_max=dataset.max()+2
x_axis=np.linspace(x_min,x_max,100)
bandwith =((4*dataset.std()**5)/(3*len(dataset)))**0.2

kanel_list =[]
for data_point in dataset:
    kanel =stats.norm(data_point,bandwith).pdf(x_axis)
    kanel_list.append(kanel)
    kanel=kanel/kanel.max()
    kanel =kanel*0.4
    plt.plot(x_axis,kanel,color="red",alpha=0.5)

plt.ylim(0.1)
```

```python
sns.rugplot(dataset);

# X軸を競ってい
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 全体を等間隔に100分割します。
x_axis = np.linspace(x_min,x_max,100)

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2

kernel_list = []

for data_point in dataset:
    
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)

    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)

```

```python
sum_of_kde =np.sum(kanel_list,axis=0)
flg =plt.plot(x_axis,sum_of_kde,color="red",alpha=0.5)
sns.rugplot(dataset)
plt.yticks([])
plt.suptitle("Sum of the Basic Functions")
```
